# Cryptocurrency Price Forecasting with ARIMA Model

**Description:** In this project, we aim to create a robust forecasting model for predicting the prices of a selected cryptocurrency. We will leverage the AutoRegressive Integrated Moving Average (ARIMA) model, a powerful time series analysis technique, to capture temporal patterns (recurring trends or behaviors that occur over time) in cryptocurrency price data.

# 1. Data Preprocessing

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go

import warnings
warnings.filterwarnings('ignore')
```

## Loading the Historical Data

```python
data = pd.read_csv('BTCUSDT_15m.csv', parse_dates=['Date'], index_col='Date')
data
```

## Checking the data and missing values

```python
type(data.index)
```

```python
data.dtypes
```

```python
data.isnull().sum()
```

```python
## Lets try to check the general trend of the BTCUSDT prices
## its better to see general trend with daily data

data_daily = data.resample('1D').agg({
                                    'Open'   : 'first',
                                    'High'   : 'max',
                                    'Low'    : 'min',
                                    'Close'  : 'last',
                                    'Volume' : 'sum'
                                    })
data_daily.head()
```

```python
# Create a trace for the candlestick chart
trace_candlestick = go.Candlestick(
    x=data_daily.index,
    open=data_daily['Open'],
    high=data_daily['High'],
    low=data_daily['Low'],
    close=data_daily['Close'],
    name='Candlestick'
)

# Create a trace for the trendline
trace_trendline = go.Scatter(
    x=[data_daily.index[0], data_daily.index[-1]],
    y=[data_daily['Open'].iloc[0], data_daily['Close'].iloc[-1]],
    mode='lines',
    line=dict(color='blue', width=2, dash='dash'),
    name='Trendline'
)

# Create the figure
fig = go.Figure(data=[trace_candlestick, trace_trendline])

# Set layout options
fig.update_layout(
    title='BTCUSDT Candlestick Chart with Trendline',
    xaxis_rangeslider_visible=False
)

# Show the plot
fig.show()
```

Although there are fluctuations in the price during the 6 months period, there is a clear upward trend in BTCUSDT prices. While the price is 16.500 USDT in the beginning of 2023, it eventually reached to 30.000 USDT at the end of June 2023.
